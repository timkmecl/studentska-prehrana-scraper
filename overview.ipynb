```python
import pandas as pd

import scraper
```

```python
df_new, df_old = scraper.load_data()
df = scraper.merge_data(df_new, df_old)
```

```python
df[df.city == 'LJUBLJANA'].sort_values('cena_diff_percent', ascending=False).iloc[:10]
```

```python
df_nove = df[df.cena_old.isna()]
df_stare = df[df.cena_old.notna() & df.cena.notna()]
df_ukinjene = df[df.cena.isna()]
```

```python
doplacilo_novih = df_nove.doplacilo.mean()
doplacilo_starih = df_stare.doplacilo.mean()
doplacilo_starih_old = df_stare.doplacilo_old.mean()
doplacilo_ukinjenih = df_ukinjene.doplacilo_old.mean()
print(f'AVG DOPLAČILA\nnove: {doplacilo_novih:.2f} €\nstare: {doplacilo_starih:.2f} €\nstare včasih: {doplacilo_starih_old:.2f} € \nukinjene {doplacilo_ukinjenih:.2f} €')
```

```python
n_novih = len(df_nove)
n_starih = len(df_stare)
n_ukinjenih = len(df_ukinjene)
print(f'ŠTEVILO RESTAVRACIJ\nnove: {n_novih}\nstare: {n_starih}\nukinjene: {n_ukinjenih}')
```
